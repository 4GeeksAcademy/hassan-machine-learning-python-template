```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
abnb_data = pd.read_csv('../data/raw/AB_NYC_2019.csv')
abnb_data.head(20)
```

```python

# Create a multiple histogram
sns.scatterplot(data=abnb_data, x='neighbourhood_group', y='price')


# Show the plot
plt.show()
```

```python
abnb_data["neighbourhood_group"].unique()
```

```python
sub_data = abnb_data[["neighbourhood_group", "room_type", "price"]]
sub_data
```

```python
average_prices = sub_data.groupby(['neighbourhood_group', 'room_type']).price.mean().reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a bar plot
sns.barplot(x='neighbourhood_group', y='price', hue='room_type', data=average_prices)

# Adding labels and title
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.title('Average Price by Room Type and Neighbourhood Group')
plt.xticks(rotation=45)  # Rotating x-labels for better readability if needed
plt.show()
```
